```python
from scipy import stats
from numpy import arange

threestats = pd.read_csv('threebyseason.csv')

fulldf = threestats.copy()
caccdf = threestats[(threestats['Conference'] == 'CACC') & (threestats['Tourney'] == 0)]
caccdf = caccdf.copy()
ne10df = threestats[(threestats['Conference'] == 'NE10') & (threestats['Tourney'] == 0)]
ne10df = ne10df.copy()
eccdf = threestats[(threestats['Conference'] == 'ECC') & (threestats['Tourney'] == 0)]
eccdf = eccdf.copy()

tourney = threestats[(threestats['Tourney'] == 1)]
tourney = tourney.copy()

eccdf['Winpct'] = round(((eccdf['Wins'] / (eccdf['Wins'] + eccdf['Losses']) * 100)),1)
ne10df['Winpct'] = round(((ne10df['Wins'] / (ne10df['Wins'] + ne10df['Losses']) * 100)),1)
caccdf['Winpct'] = round(((caccdf['Wins'] / (caccdf['Wins'] + caccdf['Losses']) * 100)),1)
tourney['Winpct'] = round(((tourney['Wins'] / (tourney['Wins'] + tourney['Losses']) * 100)),1)
fulldf['Winpct'] = round(((fulldf['Wins'] / (fulldf['Wins'] + fulldf['Losses']) * 100)),1)


xi = arange(10,36)
yi = arange(0,101)

slope, intercept, r_value, p_value, std_err = stats.linregress(fulldf['3PA'], fulldf['Winpct'])
line = slope*xi+intercept


trace = go.Scatter(
    x = eccdf['3PA'],
    y = eccdf['Winpct'],
    name = 'ECC',
    text = eccdf['Team'] + ' ' + eccdf['Season'],
    hoverinfo = 'text',
    mode = 'markers')

ncaa = go.Scatter(
    x = tourney['3PA'],
    y = tourney['Winpct'],
    name = 'NCAA',
    text = tourney['Team'] + ' ' + tourney['Season'],
    hoverinfo = 'text',
    mode = 'markers')

trace2 = go.Scatter(
    x = ne10df['3PA'],
    y = ne10df['Winpct'],
    name = 'NE10',
    text = ne10df['Team'] + ' ' + ne10df['Season'],
    hoverinfo = 'text',
    mode = 'markers')

trace3 = go.Scatter(
    x = caccdf['3PA'],
    y = caccdf['Winpct'],
    name = 'CACC',
    text = caccdf['Team'] + ' ' + caccdf['Season'],
    hoverinfo = 'text',
    mode = 'markers')

line = go.Scatter(
                  x=xi,
                  y=line,
                  mode='lines',
                  # marker=go.Marker(color='rgb(31, 119, 180)'),
                  name='Fit'
                  )

line2 = go.Scatter(
                  x=[20]*100,
                  y=yi,
                  mode='lines',
                  opacity = 0.2,
                  marker = dict(
                        color = 'rgb(0, 0, 0)',
                        size = .1),
                    showlegend = False,
                    name = 'Average 3PA'
                  )

line3 = go.Scatter(
                  x=xi,
                  y=[50]*25,
                  mode='lines',
                  opacity = 0.2,
                  marker = dict(
                        color = 'rgb(0, 0, 0)',
                        size = .1),
                    showlegend = False,
                    name = 'Average Record'
                  )
                
data = [trace, trace2, trace3, ncaa, line, line2, line3]

layout = dict(
    title = 'Region Win Percentage Plotted Against 3-Point Field Goal Attempts',
    hovermode = 'closest',
    yaxis=dict(
        rangemode = "tozero",
        tickmode='linear',
        ticks='outside',
        tick0=0,
        dtick=10,
        ticklen=8,
        tickwidth=1,
        tickcolor='#000',
        range=[0,100]),
    xaxis=dict(
        tickmode='linear',
        ticks='outside',
        tick0=0,
        dtick=1,
        ticklen=8,
        tickwidth=1,
        tickcolor='#000',
        range=[10,35]
    ),
    annotations=[
        dict(
                  x=15,
                  y=10,
                  text='Low 3PA / Bad Record',
                  showarrow=False,
                  font=dict(size=12)
        ),
        dict(
              x=15,
              y=90,
              text='Low 3PA / Good Record',
              showarrow=False,
              font=dict(size=12)
        ),
        dict(
              x=30,
              y=10,
              text='High 3PA / Bad Record',
              showarrow=False,
              font=dict(size=12)
        ),
        dict(
              x=30,
              y=90,
              text='High 3PA / Good Record',
              showarrow=False,
              font=dict(size=12)
        )
    ]
    )
    
fig = dict(data=data, layout=layout)
iplot(fig, filename = 'threesxwinpct')

```
